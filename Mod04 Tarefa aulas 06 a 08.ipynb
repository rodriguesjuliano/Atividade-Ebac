````python
# 4) Seu código aqui

#qtd status por ano
pd.DataFrame(df[['status','ano']].value_counts())
````

````python
# 4) Seu código aqui

#salvando o arquivo em csv

#Carregando o arquivo dos estados
url4='https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?tipo_interessado=Estado'
````

````python
#selecionando o conjunto de dados
dados_estados=pd.DataFrame(data_json_estados['items'])

#lendo as duas primeiras linhas do df
dados_estados.head(2)
````

````python
#Validando dataframe criado
list(dados_estados['tipo_interessado'].unique())
````

````python
#filtrando o estado
dados_estados_bahia=dados_estados[dados_estados['uf']=='BA']

#filtrando o tipo de status
dados_ba_status=pd.DataFrame(dados_estados_bahia[dados_estados_bahia['status']=='Deferido'])
````

````python
#salvando em csv
dados_ba_status.to_csv('solicitacoes bahia.csv',sep=';',index=False,encoding='utf8',decimal=',')
````
