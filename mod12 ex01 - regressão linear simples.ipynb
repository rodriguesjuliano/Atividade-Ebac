````python
import pandas as pd
import seaborn as sns
from seaborn import load_dataset
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.formula.api as smf
````

### 1. Explorando a base de gorjetas

a. Na aula fizemos uma regressão de ```tip``` explicada por ```net_bill```. Carregue os pacotes e a base, crie as variáveis necessárias como na aula, e reproduza esta regressão.

b. Reproduza também o gráfico de dispersão dessas duas variáveis.

````python
#importando base de gorjetas

tips=sns.load_dataset("tips")
#percentual da gorjeta em relação a toda a conta
tips["tip_pct"]=tips["tip"]/(tips["total_bill"]-tips["tip"])
#valor da conta sem a gorjeta
tips["net_bill"]=tips["total_bill"]-tips["tip"]
tips.head()
````

````python
#Gráfico de dispersão do seaborn, retirando o soutliers identificados na aula passada (data<.5)
sns.regplot(y='tip',x='net_bill',data=tips[tips['tip_pct']<.5])

#gerando linha de referencia de 10%
x=np.linspace(0,tips['net_bill'].max(),100)
y=.1*x

#plotando linha dos 10% (em vermelho)
plt.plot(x,y,'-r',label='linha dos 10%')

#opção estética para aparecer o (0,0) no gráfico
plt.xlim(left=0)
plt.ylim(bottom=0)
````

````python
#guardando todos os resultados do modelo de regressão linear
#função ols=função do pacote formula do statesmodel
#tip (variável resposta)
#variável explicativa é o valor da conta (net_bill)
#data= base de dados
#pode-se colocar mais variáveis explicativas com a sintaxe "net_bill + smoke + size"
#fit= ajustar a regressão aos dados da tabela

reg=smf.ols('tip ~ net_bill',data=tips).fit()
reg.summary()
````
